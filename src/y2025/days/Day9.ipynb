```python
import matplotlib.pyplot as plt
import os

with open('../inputs/day9.txt', 'r') as f:
    coords = [list(map(int, line.split(','))) for line in f]

x_coords = [coord[0] for coord in coords]
y_coords = [coord[1] for coord in coords]

x_coords.append(coords[0][0])  # close the loop by adding the first point at the end
y_coords.append(coords[0][1])  # close the loop by adding the first point

plt.figure(figsize=(8, 6))
plt.plot(x_coords, y_coords, linestyle='-', color='b')  

plt.title('Connected Dots')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.grid(True)
plt.show()
```
